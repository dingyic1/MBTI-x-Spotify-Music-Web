```python
import math
# def MSE(row, MBTI_info):
#   MSE = (1/10) * (math.sqrt((row['danceability']-MBTI_info['danceability_mean'])**2) +
#             math.sqrt((row['energy']-MBTI_info['energy_mean'])**2) +
#             math.sqrt((row['loudness']-MBTI_info['loudness_mean'])**2) +
#             math.sqrt((row['mode']-MBTI_info['mode_mean'])**2) +
#             math.sqrt((row['speechiness']-MBTI_info['speechiness_mean'])**2) +
#             math.sqrt((row['acousticness']-MBTI_info['acousticness_mean'])**2) +
#             math.sqrt((row['instrumentalness']-MBTI_info['instrumentalness_mean'])**2) +
#             math.sqrt((row['liveness']-MBTI_info['liveness_mean'])**2) +
#             math.sqrt((row['valence']-MBTI_info['valence_mean'])**2) +
#             math.sqrt((row['tempo']-MBTI_info['tempo_mean'])**2)
#              )   
            
#               # 
#   return MSE
def MBTIDecision(row, MBTI_df):
  new_df =  MBTI_df.copy()
  new_df['MSE'] = (1/10) * (np.sqrt((row['danceability']-new_df['danceability_mean'])**2) +
            np.sqrt((row['energy']-new_df['energy_mean'])**2) +
            np.sqrt((row['loudness']-new_df['loudness_mean'])**2) +
            np.sqrt((row['mode']-new_df['mode_mean'])**2) +
            np.sqrt((row['speechiness']-new_df['speechiness_mean'])**2) +
            np.sqrt((row['acousticness']-new_df['acousticness_mean'])**2) +
            np.sqrt((row['instrumentalness']-new_df['instrumentalness_mean'])**2) +
            np.sqrt((row['liveness']-new_df['liveness_mean'])**2) +
            np.sqrt((row['valence']-new_df['valence_mean'])**2) +
            np.sqrt((row['tempo']-new_df['tempo_mean'])**2)
             )  
  MBTI_key = list(new_df[new_df.MSE == new_df.MSE.min()]['mbti'])[0]
  return MBTI_key
# new_df = final_tracks_df.copy().head(10)
# new_df['MBTI'] = new_df.apply(lambda x: MBTIDecision(x, MBTI_df_filter), axis=1)
final_tracks_df['MBTI'] = final_tracks_df.swifter.apply(lambda x: MBTIDecision(x, MBTI_df_filter), axis=1)
```
